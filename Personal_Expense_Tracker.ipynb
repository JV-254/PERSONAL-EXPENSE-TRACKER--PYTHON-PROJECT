```python
import pandas as pd
# Creating a dataframe
data={
    'Date':['2025-06-01','2025-06-01','2025-06-02','2025-06-03','2025-06-04','2025-06-05','2025-06-06','2025-06-07','2025-06-08','2025-06-09'],
    'Category':['Food','Transport','Food','Entertainment','Utilities','Food','Transport','Shopping','Food','Entertainment'],
    'Description':['Lunch','Bus Fare','Breakfast','Movie','Internet','Dinner','Taxi','T-shirt','Lunch','Games'],
    'Amount':[350,100,200,800,2500,600,500,1200,400,700],
    'Payment Method':['Cash','Mpesa','Cash','Mpesa','Mpesa','Cash','Mpesa','Card','Cash','Mpesa']
}
df=pd.DataFrame(data)
# Covert the dataframe to CSV
df.to_csv('Expense.csv')
df
```

```python
import pandas as pd
# Loading dataset into pandas
df=pd.read_csv('Expense.csv')
df
# Exploring the Dataset
# Reading the first 5 rows
ED=df.head()
print(ED)
# Exploring the last 5 rows
print('===========================================================================================')
ED1=df.tail()
print(ED1)
# Viewing the dataset information
print('===========================================================================================')
ED2=df.info()
print(ED2)

# Viewing dataset description
print('===========================================================================================')
ED3=df.describe()
print(ED3)

# Remmoving unnecessary columns
df=df.drop(columns='Unnamed: 0')
df

# Finding the shape of the dataset
ED4=df.shape
print(ED4)

# DATA ANALYSIS
# Calculating the Number of Expenses
print('===========================================================================================')
ED5=df.shape[0]
print(f'Number of Expenses: {ED5}')

# Total Money Spent
print('===========================================================================================')
ED6=df['Amount'].sum()
print(f'Total Money Spent: {ED6}')

# Average Expense
print('===========================================================================================')
ED7=df['Amount'].mean()
print(f'Average Expense: {ED7}')

# Largest Expense
print('===========================================================================================')
ED8=df['Amount'].max()
print(f'Largest Expense: {ED8}')

# Smallest Expense
print('===========================================================================================')
ED9=df['Amount'].min()
print(f'Smallest Expense: {ED9}')

# CATEGORY ANALYSIS
# Total spending per category
print('===========================================================================================')
ED10=df.groupby('Category')['Amount'].sum()
print(f'Total Spending per Category: \n {ED10}')

# Highest Spending Category
print('===========================================================================================')
ED11=df.groupby('Category')['Amount'].sum().idxmax()
print(f'Highest Spending Category: \n {ED11}')

# FILTERING DATA
# All Food Expenses
print('===========================================================================================')
ED12=df[df['Category']=='Food']
print(f'All Food Expenses: \n {ED12}')

# Expenses above 1000
print('===========================================================================================')
ED13=df[df['Amount']>1000]
print(f'Expenses Above 1000: \n {ED13}')

# Creating a monthly Summary
# Converting the date column
print('===========================================================================================')
df['Date']=pd.to_datetime(df['Date'])
print(df.info())

# Extracting the month
df['Month']=df['Date'].dt.month
df

# Extracting the month name
df['Month Name']=df['Date'].dt.month_name()
df

# Extracting the month name first three letters
df['Month Name']=df['Date'].dt.strftime('%b')
df

# Extracting the Year
df['Year']=df['Date'].dt.year
df

# Changing the date format
df['Date1']=df['Date'].dt.strftime('%d-%m-%Y')
df
# Time Series Analysis
# Total Spendings per month
print('===========================================================================================')
ED14=df.groupby('Month Name')['Amount'].sum()
print(f'Total Spendings Per Month: \n {ED14}')

# Expenses in each category
print('===========================================================================================')
ED15=df.groupby('Category')['Amount'].count()
print(f'Expenses in each Category: \n {ED15}')

# Top three most expensive Transactions
print('===========================================================================================')
ED16=df.nlargest(3,'Amount')
print(f'Top 3 Most Expensive Transactions: \n {ED16}')

# Percentage of spending of each category
print('===========================================================================================')
ED17=df.groupby('Category')['Amount'].sum()
ED18=df['Amount'].sum()
ED19=(ED17/ED18)*100
ED19=ED19.round(2)
print(f'Percentage of each Category: \n {ED19}')

# All Expenses paid using M pesa
print('===========================================================================================')
ED20=df[df['Payment Method']=='Mpesa']
print(f'All Expenses paid using Mpesa: \n{ED20}')

# Adding a column called Expense level
# Create a function
def Expense(level):
    if level<500:
        return 'Low'
    elif level<1000:
        return 'Medium'
    else:
        return 'High'
    
df['Expense Level']=df['Amount'].apply(Expense)
df
```
